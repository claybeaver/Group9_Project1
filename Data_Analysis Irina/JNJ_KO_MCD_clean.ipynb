```python
import pandas as pd
```

```python
jnj_path = "JNJ.csv"
mcd_path = "MCD.csv"
pfe_path = "PFE.csv"
ko_path = "KO.csv"
sbux_path = "sbux.csv"
jnj_df = pd.read_csv(jnj_path)
mcd_df = pd.read_csv(mcd_path)
pfe_df = pd.read_csv(pfe_path)
ko_df = pd.read_csv(ko_path)
sbux_df = pd.read_csv(sbux_path)
```

```python
jnj_df.tail()
```

```python
# Rename Adj Close column to Ticker Symbol for the Stock.
jnj_df = jnj_df.rename(columns = {'Adj Close': 'JNJ'})
mcd_df = mcd_df.rename(columns = {'Adj Close': 'MCD'})
ko_df = ko_df.rename(columns = {'Adj Close': 'KO'})
pfe_df = pfe_df.rename(columns = {'Adj Close': 'PFE'})
sbux_df = sbux_df.rename(columns = {'Adj Close': 'SBUX'})
```

```python
# Drop all columns except Date & Renamed Ticker Symbol column
jnj_clean = jnj_df[['Date','JNJ']]
mcd_clean = mcd_df[['Date','MCD']]
ko_clean = ko_df[['Date','KO']]
pfe_clean = pfe_df[['Date','PFE']]
sbux_clean = sbux_df[['Date','SBUX']]
sbux_clean.tail()
```

```python
merged_df = pd.merge(jnj_clean, mcd_clean, on='Date')
merged_df = pd.merge(merged_df, ko_clean, on='Date')
merged_df = pd.merge(merged_df, pfe_clean, on='Date')
merged_df = pd.merge(merged_df, sbux_clean, on='Date')
merged_df.set_index(merged_df['Date'])
merged_df
```

```python
merged_df.to_csv("JNJ_MCD_KO_PFE_SBUX_10Years.csv", encoding="utf-8", index=False)
```
