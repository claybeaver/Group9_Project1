```python
import pandas as pd
```

```python
clay_dataclean = pd.read_csv('Clay_dataclean.csv')
irina_dataclean = pd.read_csv('Irina_dataclean.csv')
lucas_dataclean = pd.read_csv('Lucas_dataclean.csv')
tunde_dataclean = pd.read_csv('Tunde_dataclean.csv')
```

```python
clay_dataclean.head()
```

```python
irina_dataclean.head()
```

```python
lucas_dataclean.head()
```

```python
lucas_dataclean = lucas_dataclean.drop(["NFLX_y"], axis=1)
lucas_dataclean.head()
```

```python
lucas_dataclean = lucas_dataclean.rename(columns = {'NFLX_x': 'NFLX'})
lucas_dataclean.head()
```

```python
tunde_dataclean.head()
```

```python
clay_irina_merge = pd.merge(clay_dataclean,irina_dataclean, on="Date")
clay_irina_merge.head()
```

```python
lucas_tunde_merge = pd.merge(lucas_dataclean,tunde_dataclean, on="Date")
lucas_tunde_merge.head()
```

```python
project_data_df = pd.merge(clay_irina_merge,lucas_tunde_merge, on="Date")
project_data_df.head()
```

```python
project_data_df.to_csv('FINAL_DF.csv', encoding="utf-8", index=False)
```
