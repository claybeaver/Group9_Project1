```python
import pandas as pd
```

```python
#my paths
T_df = pd.read_csv('T.csv')
WMT_df = pd.read_csv('WMT.csv')
TMUS_df = pd.read_csv('TMUS.csv')
COST_df = pd.read_csv('COST.csv')
NFLX_df = pd.read_csv('NFLX.csv')
DIS_df = pd.read_csv('DIS.csv')
```

```python
T_df.head()
```

```python
#my renamed tables
T_df = T_df.rename(columns = {'Adj Close': 'T'})
WMT_df = WMT_df.rename(columns = {'Adj Close': 'WMT'})
TMUS_df = TMUS_df.rename(columns = {'Adj Close': 'TMUS'})
COST_df = COST_df.rename(columns = {'Adj Close': 'COST'})
NFLX_df = NFLX_df.rename(columns = {'Adj Close': 'NFLX'})
DIS_df = DIS_df.rename(columns = {'Adj Close': 'DIS'})
```

```python
WMT_df.head()
```

```python
#my renamed and cleaned tables
T_clean = T_df[['Date','T']]
WMT_clean = WMT_df[['Date','WMT']]
TMUS_clean = TMUS_df[['Date','TMUS']]
COST_clean = COST_df[['Date','COST']]
NFLX_clean = NFLX_df[['Date','NFLX']]
DIS_clean = DIS_df[['Date','DIS']]
```

```python
TMUS_clean.head()
```

```python
len(COST_clean)
```

```python
len(NFLX_clean.dropna())
```

```python
Lucas_merge_info1 = pd.merge(T_clean, WMT_clean, on = "Date")
```

```python
Lucas_merge_info1.head()
```

```python
Lucas_merge_info2 = pd.merge(Lucas_merge_info1, TMUS_clean, on = "Date")
```

```python
Lucas_merge_info2.head()
```

```python
Lucas_merge_info3 = pd.merge(Lucas_merge_info2, COST_clean, on = "Date")
```

```python
Lucas_merge_info3.head()
```

```python
Lucas_merge_info4 = pd.merge(Lucas_merge_info3, NFLX_clean, on = "Date")
```

```python
Lucas_merge_info4.head()
```

```python
Lucas_merge_info6 = pd.merge(Lucas_merge_info4, DIS_clean, on = "Date")
```

```python
Lucas_merge_info6.head()
```

```python
Lucas_merge_table_complete = Lucas_merge_info6
Lucas_merge_table_complete.head()
```

```python
len(Lucas_merge_table_complete)
```

```python
Lucas_merge_table_complete.to_csv('Lucas_clean_data.csv', encoding="utf-8", index=False)
```
